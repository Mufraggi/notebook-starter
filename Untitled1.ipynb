```python
import pandas as pd
import matplotlib.pyplot as plt

# Chargement des données depuis un fichier CSV
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

# 1. Statistiques générales
def general_statistics(df):
    total_passengers = len(df)
    total_survivors = df['survived'].sum()
    survival_rate = total_survivors / total_passengers
    average_age = df['age'].mean()
    average_fare = df['fare'].mean()

    print(f"Total Passengers: {total_passengers}")
    print(f"Total Survivors: {total_survivors}")
    print(f"Survival Rate: {survival_rate:.2f}")
    print(f"Average Age: {average_age:.2f}")
    print(f"Average Fare: {average_fare:.2f}")

# 2. Répartition par classe
def class_distribution(df):
    class_counts = df['pclass'].value_counts()
    class_survivors = df.groupby('pclass')['survived'].sum()
    survival_rate_by_class = class_survivors / class_counts

    class_counts.plot(kind='bar', rot=0)
    plt.title("Passenger Distribution by Class")
    plt.xlabel("Class")
    plt.ylabel("Count")
    plt.show()

    survival_rate_by_class.plot(kind='bar', rot=0)
    plt.title("Survival Rate by Class")
    plt.xlabel("Class")
    plt.ylabel("Survival Rate")
    plt.show()

# 3. Répartition par sexe
def gender_distribution(df):
    gender_counts = df['sex'].value_counts()
    gender_survivors = df.groupby('sex')['survived'].sum()
    survival_rate_by_gender = gender_survivors / gender_counts

    gender_counts.plot(kind='bar', rot=0)
    plt.title("Passenger Distribution by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()

    survival_rate_by_gender.plot(kind='bar', rot=0)
    plt.title("Survival Rate by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Survival Rate")
    plt.show()

# 4. Âge et catégories d'âge
def age_distribution(df):
    df['age'].plot(kind='hist', bins=20)
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

# 5. Voyage en famille
def family_travel(df):
    with_family = df[(df['sibsp'] > 0) | (df['parch'] > 0)]
    alone = df[(df['sibsp'] == 0) & (df['parch'] == 0)]

    family_counts = pd.Series([len(with_family), len(alone)], index=['With Family', 'Alone'])
    family_counts.plot(kind='bar', rot=0)
    plt.title("Passenger Distribution - Traveling with Family or Alone")
    plt.xlabel("Category")
    plt.ylabel("Count")
    plt.show()

# 6. Embarquement
def embarkation_distribution(df):
    embark_counts = df['embarked'].value_counts()
    embark_survivors = df.groupby('embarked')['survived'].sum()
    survival_rate_by_embark = embark_survivors / embark_counts

    embark_counts.plot(kind='bar', rot=0)
    plt.title("Passenger Distribution by Embarkation Port")
    plt.xlabel("Embarkation Port")
    plt.ylabel("Count")
    plt.show()

    survival_rate_by_embark.plot(kind='bar', rot=0)
    plt.title("Survival Rate by Embarkation Port")
    plt.xlabel("Embarkation Port")
    plt.ylabel("Survival Rate")
    plt.show()

# 7. Cabines
def cabin_distribution(df):
    df['cabin'].value_counts().plot(kind='bar', figsize=(10, 5))
    plt.title("Cabin Distribution")
    plt.xlabel("Cabin")
    plt.ylabel("Count")
    plt.show()

# 8. Bateaux de sauvetage
def lifeboat_distribution(df):
    lifeboat_counts = df['boat'].value_counts().drop('')
    lifeboat_counts.plot(kind='bar', figsize=(10, 5))
    plt.title("Lifeboat Distribution")
    plt.xlabel("Lifeboat")
    plt.ylabel("Count")
    plt.show()

# 9. Identification des corps
def body_identification(df):
    body_identified = df[df['body'].notnull()]
    body_unidentified = df[df['body'].isnull()]

    body_counts = pd.Series([len(body_identified), len(body_unidentified)], index=['Body Identified', 'Body Unidentified'])
    body_counts.plot(kind='bar', rot=0)
    plt.title("Passenger Distribution - Body Identification")
    plt.xlabel("Category")
    plt.ylabel("Count")
    plt.show()

# 10. Destinations
def destination_distribution(df):
    destinations = df['home.dest'].str.split(',', expand=True)[0].str.strip()
    destination_counts = destinations.value_counts().head(10)

    destination_counts.plot(kind='bar', figsize=(12, 6))
    plt.title("Top 10 Passenger Destinations")
    plt.xlabel("Destination")
    plt.ylabel("Count")
    plt.show()


```

```python
data = load_data('titanic.csv')

general_statistics(data)
```
